```python
class node:
    
    def __init__(self,data):
        self.data = data
        self.nextnode = None
        
class linkedlist:
    
    def __init__(self,head=None):
        
        self.head = head
        self.size = 0
        
    def traverse_linkedlist(self):
        
        self.length = 0
        curr_node = self.head
        if self.head:
            self.length += 1
            while curr_node.nextnode != None:
                self.length +=1
                curr_node = curr_node.nextnode
            return self.length
    
    def find_data_in_linkedlist(self,data):
        curr_node = self.head
        self.data = data
        if self.head:
            while curr_node != None:
                if curr_node.data == self.data:
                    return True
                else:
                    curr_node = curr_node.nextnode
        else:
            return False
    
    def add_node_at_the_beginning(self,data):
        
        newnode = node(data)
        newnode.nextnode = self.head
        self.head = newnode
        self.size += 1
        return True
    
    def add_node_at_the_end(self,data):
        
        newnode = node(data)
        curr_node = self.head
        if self.head:
            while curr_node.nextnode != None:
                curr_node = curr_node.nextnode
            curr_node.nextnode = newnode
            newnode.nextnode = None
            self.size += 1
        else:
            print("Empty List...")
        return True
    
    def add_node_at_given_ind(self,data,GIndex):
        
        self.GIndex = GIndex
        newnode = node(data)
        curr_node = self.head
        SIndex = 0
        
        if self.GIndex == 0:
            
            newnode.nextnode = curr_node
            self.head = newnode
            self.size += 1
        else:
            if self.GIndex <= self.size:
                while self.GIndex != SIndex:
                    prev_node = curr_node
                    curr_node = curr_node.nextnode
                    SIndex += 1
                prev_node.nextnode = newnode
                newnode.nextnode = curr_node
                self.size += 1
            else:
                print("Invalid Index..")
        return True


    
    ## Remove duplicates from a sorted linked list  
    def remove_duplicates(self):
        
        curr_node = self.head
        prev_node = None
        Next = None
        while curr_node.nextnode != None:
            Next = curr_node.nextnode
            if curr_node.data == Next.data:
                curr_node.nextnode = Next.nextnode
                Next = Next.nextnode
            else:
                curr_node = curr_node.nextnode
                
    ### Find nth node from the end of the linked list
    
    def find_nth_node_from_end(self,n):
        
        self.n = n
        curr_node = self.head
        temp = self.head
        for i in range(n-1):
            temp = temp.nextnode
        print("temp:",temp.data)
        while temp.nextnode != None:
            temp = temp.nextnode
            curr_node = curr_node.nextnode
        return curr_node.data
    
    ##Move last node to the front in a given linked list
    
    def move_last_node_to_front(self):
        
        curr_node = self.head
        temp = self.head
        while curr_node.nextnode != None:
            
            prev_node = curr_node
            curr_node = curr_node.nextnode
        prev_node.nextnode = None
        curr_node.nextnode = self.head
        self.head = curr_node
        
    ### Group all odd nodes together followed by even nodes. 
    def move_odd_even(self):
        #odd pointer
        oddpointer = self.head
        #even pointer
        evenpointer = self.head.nextnode
        #To preserver even pointer head
        evenhead = evenpointer
        while evenpointer !=None and evenpointer.nextnode != None:
            
            oddpointer.nextnode = evenpointer.nextnode
            oddpointer = oddpointer.nextnode
            evenpointer.nextnode = oddpointer.nextnode
            evenpointer = evenpointer.nextnode
        oddpointer.nextnode = evenhead
        ##not sure why head is being returned
        return self.head
    ### Group all even nodes together followed by odd nodes.
    def move_even_odd(self):
        ##odd pointer:od
        od = self.head
        ##even pointer : ev
        ev = self.head.nextnode
        ##odd head:oh
        oh = od
        ##even head:eh
        eh = ev
        while ev !=None and ev.nextnode != None:
            
            od.nextnode = od.nextnode.nextnode
            ev.nextnode = ev.nextnode.nextnode
            od = od.nextnode
            prev = ev
            ev = ev.nextnode
            if ev == None:
                ev = prev

        ev.nextnode = oh
        od.nextnode = None
        self.head = eh
        
    ### Group all odd nodes together followed by even nodes in reverse order. 
    def move_odd_even_in_reverse(self):
        #odd pointer
        od = self.head
        #even pointer
        ev= self.head.nextnode
        #To preserver even pointer head
        eh = ev
        while ev !=None and ev.nextnode != None:
            
            od.nextnode = ev.nextnode
            od = od.nextnode
            ev.nextnode = od.nextnode
            ev = ev.nextnode
       
        od.nextnode = eh
        ev = od.nextnode
        while ev.nextnode !=None:
            Next = ev.nextnode
            ev.nextnode = Next.nextnode
            Next.nextnode = od.nextnode
            od.nextnode = Next
            
            
    ### Reverse linked list in pair 
    def reverse_in_pair(self):
        
        p = self.head
        sh = p.nextnode
        dummy = linkedlist(0)
        dummy.nextnode = self.head
        while dummy.nextnode and dummy.nextnode.nextnode != None:
            p = dummy.nextnode
            q = p.nextnode
            dummy.nextnode = q
            p.nextnode = q.nextnode
            q.nextnode = p
            dummy = p
            
        self.head = sh
        
   ### Reverse nodes in a group of size = k
   ##### some work still needed
    def reverse_nodes_in_group(self,k):
        
        self.k = k
        curr = self.head
        p = self.head
        tail = None
        prev = linkedlist(0)
        prev.next = self.head
        x = self.traverse_linkedlist()
        print("length:",x)
        for i in range(k-1):
            finalhead = curr.nextnode
            curr = curr.nextnode
       
        curr = self.head
       
        while curr and curr.nextnode != None:
            
            for j in range(k-1):
                newnode = p.nextnode
                p = p.nextnode
            p = p.nextnode
            head = curr
            while curr != p:
                Next = curr.nextnode
                curr.nextnode = prev
                prev = curr
                curr = Next

            if tail != None:
                tail.nextnode = prev

            tail = head
            head = curr   
            prev = None
            x = x - k
            
            

        self.head = finalhead

   
   
           
         
    ### Display Linked List from the end
    def display_linked_from_the_end(self):
        
        curr_node = self.head
        self.display_recur(curr_node)
        
    def display_recur(self,curr_node):
        if curr_node==None:
            return
        
        self.display_recur(curr_node.nextnode)
        print(curr_node.data,end=" , ")
 
                   
    def print_single_linked_list(self):
        
        curr_node = self.head
        print(curr_node.data)
        mylist = []
        if self.head:
            
            while curr_node!= None:
                
                mylist.append(curr_node.data)
                curr_node = curr_node.nextnode
            print(mylist)
            print("Size of the list:",self.traverse_linkedlist())
        else:
            print("The linked list is empty..")
            
l = linkedlist()
#l.add_node_at_the_beginning('bb')
#l.add_node_at_the_beginning("ggo")
#l.add_node_at_the_beginning('a')
l.add_node_at_given_ind('a',0)
l.add_node_at_given_ind('ab',1)
l.add_node_at_given_ind('abc',2)
l.add_node_at_given_ind('d',3)
l.add_node_at_given_ind('e',4)
l.add_node_at_given_ind('qr',5)
l.add_node_at_given_ind('wxy',6)
l.add_node_at_given_ind('qrs',7)
l.add_node_at_given_ind('ff',8)
l.print_single_linked_list()

```
