```python
## Graph Depth First Search using Adjancency List(Start -- May16th,2021)
## DFS for connected graph.(Start:May16th,2021---End:May17th,2021)
## DFS for disconnected graph.(Start:May16th,2021---End:May17th,2021)

class Graph:
    
    def __init__(self):
        self.graph = {}
        self.vertices = 0
        
    def add_vertex(self,vertex):
        
        if vertex in self.graph:
            print("Vertex",vertex,"already exists..")
        else:
            self.graph[vertex] = []
            self.vertices += 1
            
            
    def add_edge(self,v1,v2,weight):
        
        if v1 not in self.graph:
            print("Vertex",v1,"does not exist...")
        elif v2 not in self.graph:
            print("Vertex",v2,"does not exist...")
        else:
            self.graph[v1].append([v2,weight])
            
    def remove_edge(self,vertex):
        
        if vertex not in self.graph:
            print("Vertex",vertex,"does not exist...")
            
        else:
            for key in list(self.graph.keys()):
                if key == vertex:
                    self.graph.pop(key)
                else:
                    for val in self.graph[key]:
                        if val[0] == vertex:
                            self.graph[key].remove(val)
                            
    def print_graph(self):
        print("The given graph is...")
    
        for k,v in self.graph.items():
            print(k,"-->",v)
            
    def DFS_For_Connected_Graph(self,start_vertex):
        
        visited = []
        visited.append(start_vertex)
        ##Initializing stack
        s = []
        s.append(start_vertex)
        print("The depth first traversal for the given graph is:")
        while len(s) > 0:
            
            temp = s.pop(-1)
            print(temp,end=" ")
            for adj_vertex in self.graph[temp]:
                #print("adjacent vertex:",adj_vertex[0])
                if adj_vertex[0] not in visited:
                    visited.append(adj_vertex[0])
                    s.append(adj_vertex[0])
                    
        del visited
                    
    def DFS_Recursion(self,start_vertex):
        
        visited = []
        self.DFS_Recursion_Utility(self.graph,visited,start_vertex)
        
    def DFS_Recursion_Utility(self,graph,visited,start_vertex):
        
        if start_vertex not in visited:
            print(start_vertex,end=" ")
            visited.append(start_vertex)
            for adj_vertex in self.graph[start_vertex]:
                self.DFS_Recursion_Utility(self.graph,visited,adj_vertex[0])
                
g = Graph()
#g.add_vertex("A")
#g.add_vertex("B")
#g.add_vertex("C")
#g.add_vertex("D")
#g.add_vertex("E")

#g.add_edge("A","B",1)
#g.add_edge("A","C",2)
#g.add_edge("B","D",3)
#g.add_edge("D","C",1)
#g.add_edge("C","E",2)
#g.add_edge("E","A",1)

g.add_vertex(1)
g.add_vertex(2)
g.add_vertex(3)
g.add_vertex(4)
g.add_vertex(5)
g.add_vertex(6)
g.add_vertex(7)

g.add_edge(1,2,1)
g.add_edge(1,3,2)
g.add_edge(2,4,1)
g.add_edge(2,5,2)
g.add_edge(3,5,1)
g.add_edge(4,6,1)
g.add_edge(5,6,1)
g.add_edge(6,7,2)
g.print_graph()
print("\n")
#g.DFS_For_Connected_Graph(1)
print("\n")
g.DFS_Recursion(1)

#{1: [2, 3], 2: [4, 5],3: [5], 4: [6], 5: [6],6: [7], 7: []}
```
