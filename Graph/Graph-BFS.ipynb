```python
## Graph Breadth First Search using Adjancency List(Start -- May16th,2021)
## BFS for connected graph.(Start:May16th,2021---End:May17th,2021)
## BFS for disconnected graph.(Start:May16th,2021---End:May17th,2021)

def add_vertex(v):
    global graph
    global vertices_no
    if v in graph:
        print("Node",v, "already exists in graph..")
    else:
        vertices_no += 1
        graph[v]=[]
    
def add_edge(v1,v2,weight):
    
    global graph
    if v1 not in graph:
        print("The vertex",v1,"is not present..")
    elif v2 not in graph:
        print("The vertex:",v2,"is not present..")
    else:
        temp = [v2,weight]
        graph[v1].append(temp)
        
def remove_edge(vertex):
    
    if vertex not in graph:
        print("Vertex does not exist..")
    else:
        
        for key in list(graph.keys()):
            
            if key == vertex:
                graph.pop(vertex)
            else:
                
                for item in graph[key]:
                    print("item:",item)
                    if item[0] ==  vertex:
                        graph[key].remove(item)

def print_graph(graph):
    
    print("The vertices are:",list(graph.keys()))
    for key,val in graph.items():
        print(key,"-->",graph[key])
        
def BFS_Connected_Graph(graph,start_vertex):
    
    visited = []
    
    ##Initializing queue
    q = []
    q.append(start_vertex)
    visited.append(start_vertex)
    
    print("The BFS traversal for the given graph is:")
    while len(q) > 0:
        
        temp = q.pop(0)
        print(temp,end=" ")
        for val in graph[temp]:
            
            if val[0] not in visited:
                visited.append(val[0])
                q.append(val[0])
             
                    
                    
def BFS_Disconnected_Graph(graph):
    
    visited = []
    
    ##Initializing queue
    q = []
    print("The BFS traversal for the given diconnected graph is:")
    
    for key in graph:
        
        if key not in visited:
            visited.append(key)
            q.append(key)
            
        while len(q) > 0:
            
            temp = q.pop(0)
            print(temp,end=" ")
            for val in graph[temp]:
                if val[0] not in visited:
                    visited.append(val[0])
                    q.append(val[0])        

    
    
graph = {} 
vertices_no = 0

add_vertex("A")
add_vertex("B")
add_vertex("C")
add_vertex("D")
add_vertex("E")
add_vertex("F")

add_edge("A","B",1)
add_edge("A","D",3)
#add_edge("D","C",2)
add_edge("B","D",2)
add_edge("C","E",1)
add_edge("E","F",1)
#add_edge("B","D",1)

print_graph(graph)
#remove_edge("B")
print("\n")
#BFS(graph,"A")
BFS_Disconnected_Graph(graph)
```
