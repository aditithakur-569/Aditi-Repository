```python
#Graph can be represented in several ways.The two most common methods are:
#1) Adjacency List
#2) Adjacency Matrix
#3) Incidence Matrix

###Advantages of Adjency List####

#1) Saves Space O(|V| + |E|) -->V(Vertices) And E(Edges)
#2) Adding an Edge takes O(1) time.
#3) Adding a vertex takes O(1) time.

### Disadvantages of Adjency List####

#1) Time complexity to find if an edge exists between vertex v and vertex u,will be O(V).
#2) Time taken to remove an edge takes O(E) time.
#3) Time taken to remove a vertex O(V+E) time.

###Advantages of Adjency Matrix####

#1) Adding/Removing an edge takes O(1) time.
#2) Time complexity to find if an edge exists between vertex v and vertex u will be O(1).

###Disadvantages of Adjency Matrix####

#1)Consumes more space, even if the graph is sparse it will take O(V^2) space.
#2) Adding/Removing vertex is O(V^2).
```

```python
## Suppose we have a list of airline tickets represented by pairs of departure and arrival airports like
##[from, to], we have to reconstruct the itinerary in order. All of the tickets belong to a man who departs from
##JFK. So, the itinerary must begin with JFK.  [start-end:May 18th,2021]
## LeetCode Problem #332
##Algorithm
## First create adjacency list for storing tickets.
## Then sort the "items" for all the key of the adjacency list.
## Perform DFS trasversal on the graph starting at "JFK".
## Reverse the array where you have stored the output.

#from collections import OrderedDict
def reconstruct_itenary(tickets):
    # to store the result
    result = []
    #stack = 
    s = []
    #adjancency list to store tickets
    graph = {}
    
    for i in tickets:
        if i[0] in graph:
            graph[i[0]].append(i[1])
        else:
            graph[i[0]] = [i[1]]
    
    for key in graph:
        x = sorted(graph[key])
        graph[key] = x 
    DFS(graph,"JFK",result)
    return result[::-1]
    
        
def DFS(graph,start_vertex,result):
    
    for adj_vertex in graph[start_vertex]:
        #print("graph before:",graph)
        #print("adj_vertex:",adj_vertex,"start_vertex:",start_vertex)
        graph[start_vertex].remove(adj_vertex)
        #print("graph after popping:",graph)
        #print("\n")
        DFS(graph,adj_vertex,result)
    result.append(start_vertex)
    

reconstruct_itenary([["JFK","SFO"],["JFK","ATL"],["SFO","ATL"],["ATL","JFK"],["ATL","SFO"]])

```
