```python
###Singly Linked List problems that uses stack
class stack():
    
    def __init__(self):
        
        self.items = []
        
    def push(self,data):
        self.items.append(data)
    
    def pop_item(self):
        return self.items.pop()
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]


#def reverse_linkedlist(self):
class node:
    
    def __init__(self,data= None,next = None):
        self.data = data
        self.next = next
        
class linkedlist:
    
    def __init__(self,head=None):
        
        self.head = head
        self.size = 0
        
    def add_node_at_the_beginning(self,data):
        
        newnode = node(data)
        newnode.next = self.head
        self.head = newnode
        self.size += 1
        return True
    
    def reverse_linked_list_using_stack(self):
        
        st = stack()
        curr = self.head
        while curr != None:
            
            st.push(curr)
            curr = curr.next
           
        newhead = st.peek()
        self.head = newhead
        curr = newhead
        st.pop_item()
        
        while st.is_empty()== False:
            
            curr.next = st.peek()
            st.pop_item()
            curr = curr.next

        curr.next = None
        
    def display_linked_list_items_from_end(self):
        
        st = stack()
        curr = self.head
        while curr:
            
            st.push(curr)
            curr = curr.next
        print(st.is_empty())
        while st.is_empty() == False:
            temp = st.peek()
            print(temp.data,end="->")
            st.pop_item()
        print("None")
                    
    
    def printList(self):
        ptr = self.head
        while ptr:
            print(ptr.data, end=" -> ")
            ptr = ptr.next
            #print(ptr)
        
        print("None")
        
l = linkedlist()
l.add_node_at_the_beginning("a")
l.add_node_at_the_beginning("d")
l.add_node_at_the_beginning("i")
l.add_node_at_the_beginning("d")
l.add_node_at_the_beginning("a")
l.add_node_at_the_beginning("s")
l.printList()
```
