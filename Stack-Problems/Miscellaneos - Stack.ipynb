```python
###Find maximum depth of nested parenthesis in a string
### Solution 1: Using Auxillary Space
## Time Complexity : O(n), Auxiliary Space : O(n)

def max_depth_nested_paren(str1):
    
    depth = 0
    count = 0
    
    for i in range(0,len(str1)):
        
        if str1[i] == "(":
            
            count += 1
            
            if count > depth:
                depth = count
        elif str1[i] == ")":
            if count > 0:
                count -= 1
            else:
                return -1
            
    if count != 0:
        return -1
        
    return depth
            
max_depth_nested_paren("( a(b) (c) (d(e(f)g)h) I (j(k)l)m)")            
```

```python
### Celebrity Problem
class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
        
    def pop_item(self):
        return self.items.pop()
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)
    
    def find_celebrity(self,arr):
        
        s = Stack()
        for i in range(0,len(arr)):
            s.push(i)
        
        while s.stack_size() >=2:
            
            i = s.pop_item()
            j = s.pop_item()
            
            if arr[i][j]==0:
                s.push(i)
            else:
                s.push(j)
        ## Potential Celebrity
        pot = s.pop_item()
        for k in range(0,len(arr)):
            if k != pot:
                if arr[k][pot] == 0 or arr[pot][k] == 1:
                    print("No Celebrity Found.....")
                    return
            else:
                continue
        return pot
            
        
    
s = Stack()
s.find_celebrity([[0,1,1,1],[1,0,1,1],[1,1,0,1],[0,0,0,0]])
```

```python
### Find maximum sum possible equal sum of three stacks...
## The solution is do with the arrays which represents stack.
class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
        
    def pop_item(self):
        return self.items.pop()
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)
    
    def max_equal_sum_of_three_stacks(self,s1,s2,s3,l1,l2,l3):
        
        sum1 = 0
        sum2 = 0
        sum3 = 0
        for i in range(0,l1):
            sum1 = sum1 + s1[i]
        #print("sum1:",sum1)
            
        for j in range(0,l2):
            sum2 = sum2 + s2[j]
        #print("sum2:",sum2)
            
        for k in range(0,l3):
            sum3 = sum3 + s3[k]
        #print("sum3:",sum3)
        
        top1,top2,top3 = 0,0,0
        
        while True:
            if top1 == l1 or top2 == l2 or top3 == l3:
                return 0
            
            if sum1 == sum2 == sum3:
                return sum1
            else:
                if sum1 >= sum2 and sum1 >= sum3:
                    sum1 = sum1 - s1[top1]
                    top1 += 1 
                
                elif sum2 >= sum1 and sum2 >= sum3:
                        sum2 = sum2 - s2[top2]
                        top2 += 1

                elif sum3 >= sum1 and sum3 >= sum2:
                    sum3 = sum3 - s3[top3]
                    top3 += 1                    
        
          
        
s = Stack()
s1 = [3,2,1,1,1]
s2 = [4,3,2]
s3 = [1,1,4,1]

l1,l2,l3 = len(s1), len(s2),len(s3)

s.max_equal_sum_of_three_stacks(s1,s2,s3,l1,l2,l3)
```

```python
### Count natural numbers whose all permutation are greater than that number
def countNumber(n): 
    result = 0
  
    # Pushing 1 to 9 because all number  
    # from 1 to 9 have this property. 
    for i in range(1, 10): 
        s = []  
        if (i <= n): 
            s.append(i)  
            result += 1
  
        # take a number from stack and add  
        # a digit smaller than last digit  
        # of it.
        print("result:",result,"s:",s)
        while len(s) != 0: 
            tp = s[-1]  
            s.pop() 
            #print("tp:",tp)
            for j in range(tp % 10, 10): 
                
                x = tp * 10 + j 
                #print("x:",x)
                if (x <= n): 
                    s.append(x)  
                    result += 1
                    print("s:",s)
  
    return result 
  
# Driver Code 
if __name__ == '__main__': 
  
    n = 15
    print(countNumber(n)) 
```
