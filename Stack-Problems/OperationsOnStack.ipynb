```python
####Reverse a stack using recursion without using loop constructs like while... for..You can only use the 
##ADT functions(isEmpty(S), push(S),pop(S)) on Stack S:


class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
    
        
    def pop_item(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return -1 
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)
        
        
    def reverse_stack(self,mystack):
        
        if mystack.is_empty():
            return
        x = mystack.pop_item()
        self.reverse_stack(mystack)
        self.insert_data(mystack,x)
        return mystack.items
    
    def insert_data(self,mystack,x):
        var = x
        #print("stack",mystack.items,"x:",x,"var:",var)

        if mystack.is_empty():
            mystack.push(x)
            #print("STACK::",mystack.items)
            return 
            
        temp = mystack.pop_item()
        #print("stack",mystack.items,"temp:",temp)
        self.insert_data(mystack,var)
        mystack.push(temp)    
    

s = Stack()
s.push("w")
s.push("o")
s.push("r")
s.push("l")
s.push("d")

s.reverse_stack(s)
```

```python
## Sort(ascending order) a Stack using recursion
## Given a stack, sort it using recursion. Use of any loop constructs like while, for..etc is not allowed. 
## We can only use the following ADT functions on Stack S.

class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
    
        
    def pop_item(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return -1 
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)
    
    def sort_stack_recursive(self,st):
        
        if st.is_empty():
            return
        
        temp = st.pop_item()
        self.sort_stack_recursive(st)
        self.sort_stack(st,temp)
        return st.items
        
    def sort_stack(self,st,temp):
        
        #print("temp:",temp,"peek:",st.peek())
        
        if st.is_empty():
            st.push(temp)
            #print("if - st:",st.items)
            return
         
        else:
            if temp < st.peek() and not st.is_empty():
                var = st.pop_item()
                self.sort_stack(st,temp)
                st.push(var)
                #print("elif - st:",st.items)
            else:
                st.push(temp)
                #print("else - st:",st.items)
            
        
s = Stack()
s.push(21)
s.push(-3)
s.push(100)
s.push(14)
s.push(9) 
s.push(2)
s.push(222)

s.sort_stack_recursive(s)

```

```python
##Sort a stack with the help of another stack

class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
    
        
    def pop_item(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return -1 
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)
    
    def sort_stack(self,st):
        
        ts = Stack()
        ts.push(st.pop_item())
        
        while not st.is_empty():
            
            x = st.pop_item()
                        
            if x >= ts.peek():
                ts.push(x)
            else:
                while not ts.is_empty() and x < ts.peek():
                    
                    y = ts.pop_item()
                    st.push(y)  
                ts.push(x)
                
        return ts.items

s = Stack()
s.push(34)
s.push(3)
s.push(31)
s.push(98)
s.push(92) 
s.push(-23)
s.sort_stack(s)
```

```python
## Given an index, delete array elements to the left which are smaller than the number at the right until the index
## is reached. Delete a[i], is a[i] < a[i+1]....

class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
    
        
    def pop_item(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return -1 
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)

    
    def delete_smaller_elements(self,arr,n):
        
        s = Stack()
        s.push(arr[0])
        count = 0
        for i in range(1,len(arr)):
            print("stack-top:",s.peek(),"arr[i]:",arr[i],"count:",count,"N:",n,"stack:",s.items)
            while not s.is_empty() and s.peek() <= arr[i] and count != n :
                s.pop_item()
                count += 1
            s.push(arr[i])     
                
        return s.items
            
        
s = Stack()
s.push(34)
s.push(3)
s.push(31)
s.push(98)
s.push(92) 
s.push(23)
s.delete_smaller_elements([34,2,16,44,98,92,23],5)

        
```

```python
## Reverse individual words in a string
## Given a string str, we need to print reverse of individual words.
class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
    
        
    def pop_item(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return -1 
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)
    
    def reverse_individual_words(self,str1):
        
        s = Stack()
        x = ""
        for i in range(0,len(str1)):
            if str1[i] != " ":
                s.push(str1[i])
            else:
                while not s.is_empty():
                    
                    x += s.pop_item()
                x += " "
        while not s.is_empty():
            x += s.pop_item()
        return x
            
s = Stack()
s.reverse_individual_words("hello World! My name is hope..")
```
