```python
### Implement two stacks in one array....
### Space Efficient Implementation

class twoStacks():
    
    def __init__(self,n):
        self.size = n
        self.arr = [None] * n
        self.stack1_index = -1
        self.stack2_index = self.size
        
    def push1(self,data):
        
        if self.stack1_index < self.stack2_index-1:
            self.stack1_index += 1
            #print("stack1_index:",self.stack1_index)
            self.arr[self.stack1_index] = data
        else:
            print("stack overflow...")
            
    def push2(self,data):
        
        if self.stack1_index < self.stack2_index-1:
            self.stack2_index -= 1
            #print("stack2_index:",self.stack2_index)
            self.arr[self.stack2_index] = data
        else:
            print("stack overflow...")
            
    def pop1(self):
        if self.stack1_index > 0:
            x = self.arr[self.stack1_index]
            self.arr.pop(self.stack1_index)
            self.stack1_index -= 1
            return x
        else:
            print("stack underflow....")
    
    def pop2(self):
        if self.stack2_index < self.size:
            y = self.arr[self.stack2_index]
            self.arr.pop(self.stack2_index)
            return y
        else:
            print("stack underflow....")
            
        
tw = twoStacks(6)
tw.push1(4)
tw.push1(9)
tw.push2(40)
tw.push2(17)
tw.push1(29)
tw.push1(67)
tw.push1(100)
print(tw.arr)
tw.pop1() 
print(tw.arr)
tw.pop2()
print(tw.arr)
```

```python
### Implement k stacks in one array in most efficient way....
### Implementation of kStacks should use only one array, 
## i.e.,k stacks should use the same array for storing elements. Following functions must be supported by kStacks.
## push(int x, int sn) –> pushes x to stack number ‘sn’ where sn is from 0 to k-1
## pop(int sn) –> pops an element from stack number ‘sn’ where sn is from 0 to k-1
## Algo for Push: 
#1. First store the element at index free in the main array(self,arr).
#2. Store "pre_free" from array Top at index st_in(which is basically stack number.
#3. Update the value of top at index st_in to free.
#4. Store the value of Next at the index free in the variable temp.
#5. Update the value of array Next at index "free" to pre_free.
#6. Finally update the new value of free to temp.
# Algo for Pop:

class kStacks:
    
    def __init__(self,N,arrsize):
        
        # No of stacks
        self.N = N
        # Size of array in which the stack items will be used to store
        self.arrsize = arrsize
        self.arr = [0] * arrsize
        self.top = [-1] * N
        self.Next = [None] * arrsize
       
        for i in range(0,arrsize):
            self.Next[i] = i+1
        self.Next[i] = -1
        self.free = 0
        
    def is_stack_free(self):
        return self.free == -1
    
    def is_Empty(self,st_in):
        return self.top[st_in] == -1
        
    def push_data(self,st_in,data):
        
        if self.is_stack_free():
            print("Stack Overflow....")
            return
  
        self.arr[self.free] = data
        print("main array:",self.arr)
        self.pre_free = self.top[st_in]
        print("pre-free:",self.pre_free)
        self.top[st_in] = self.free
        print("self.top:",self.top)
        self.temp = self.Next[self.free]
        print("temp:",self.temp)
        self.Next[self.free] = self.pre_free
        print("Next:",self.Next)
        self.free = self.temp
        print("free:",self.free)

        
    def pop(self,st_in):
        
        if self.is_Empty(st_in):
            print("Empty Stack...")
            return

        self.get_index = self.top[st_in]
        self.item = self.arr[st_in] 
        self.top[st_in] = self.Next[self.get_index]
        self.Next[self.get_index] = self.free
        self.free = self.get_index
        return self.item
    
    def display_individual_stack(self,st_in):
        
        self.temp = []
        item_ind = self.top[st_in]
        self.temp.append(self.arr[item_ind])
        #print(self.arr[item_ind],end = ",")
        while self.Next[item_ind] != -1:
            
            next_item_ind = self.Next[item_ind]
            self.temp.append(self.arr[next_item_ind])
            item_ind = next_item_ind
        return (self.temp)
        
        
k = kStacks(3,6)
k.push_data(1,22)
k.push_data(1,17)
k.push_data(0,33)
k.push_data(2,87)
x = k.pop(1)
print("The item popped was:",x)

k.push_data(2,89)
k.push_data(0,98)
k.push_data(0,99)
k.push_data(0,100)
print(k.arr)
#print(k.Next)
#print(k.top)

#x = k.pop(0)
#print("The item popped was:",x)
#x = k.pop(0)
#print("The item popped was:",x)
#x = k.pop(0)
#print("The item popped was:",x)

k.display_individual_stack(2)
```

```python
### Implement Stack using two Queues.
## There are two approaches to this program. 
## 1) Push Operation Costly :
##      - - >Push Operation Time Complexity is:O(n) and pop operation Time Complexity is: O(1)
## 2) Pop Operation Costly :
##      - - >Pop Operation Time Complexity is:O(n) and push operation Time Complexity is: O(1)
## In real life, which approach should be chosen depends on the number of operations. Ex: if we have requirement
## where we are doing 70% push and 30%pop, then we will choose Push Operation Costly approach.

##Push Operation Costly

class Queue():
    
    def __init__(self):
        self.array = []
    
    def is_empty(self):
        return self.array == []
    
    def enqueue(self,data):
        self.array.append(data)
        
    def dequeue(self):
        return self.array.pop(0)
        
class Stack():
    
    def __init__(self):
        self.queue1 = Queue()
        self.queue2 = Queue()
    
    def is_empty(self,data):
        return self.queue2.is_empty()
        
    def push_data(self,data):
        
        self.queue1.enqueue(data)
        while not self.queue2.is_empty():
                x = self.queue2.dequeue()
                self.queue1.enqueue(x)
        self.queue1,self.queue2 = self.queue2,self.queue1  
        
    def pop_data(self):
        return self.queue2.dequeue()

                
                        

s = Stack()
q = Queue()
s.push_data(5)
s.push_data(19)
s.push_data(44)
print(s.queue2.array)
#for item in s.queue2.array:
#    print(item,end=" ")
s.pop_data()
print(s.queue2.array)
```

```python
## Infix to Postfix conversion
##1. Scan the infix expression from left to right.
##2. If the scanned character is an operand, output it.
##3. Else,
##…..3.1 If the precedence of the scanned operator is greater than the precedence of the operator in the 
## stack(or the stack is empty or the stack contains a ‘(‘ ), push it.
##…..3.2 Else, Pop all the operators from the stack which are greater than or equal to in precedence than that of
## the scanned operator. After doing that Push the scanned operator to the stack. (If you encounter parenthesis
## while popping then stop there and push the scanned operator in the stack.)
##4. If the scanned character is an ‘(‘, push it to the stack.
##5. If the scanned character is an ‘)’, pop the stack and and output it until a ‘(‘ is encountered, and discard
## both the parenthesis.
##6. Repeat steps 2-6 until infix expression is scanned.
##7. Print the output
##8. Pop and output from the stack until it is not empty.

class stack():
    
    def __init__(self):
        self.items = []
        
        
    def push(self,data):
        self.items.append(data)
        
    
    def pop_item(self):
        return self.items.pop()
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
    def stack_size(self):
        
        return len(self.items)
    
    def InfixToPostfix(self,exp):
    
        prec = {}

        prec["^"] = 4
        prec["*"] = 3
        prec["/"] = 3
        prec["+"] = 2
        prec["-"] = 2
        prec["("] = 1

        s = stack()
        result = []
        for i in range(0,len(exp)):
            
            if exp[i] in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or exp[i] in "0123456789" or exp[i] in 'abcdefghijklmnopqrstuvwxyz':
                result.append(exp[i])
        
            elif exp[i] == "(":
                s.push(exp[i])
                
            elif exp[i] == ")":
                var = s.pop_item()
                while var != "(":
                    
                    result.append(var)
                    var = s.pop_item()
                
            else:
                while not s.is_empty() and prec[exp[i]] <= prec[s.peek()]:
                    result.append(s.pop_item())
                               
                s.push(exp[i])                                             
                            
        if not s.is_empty():
            while not s.is_empty():
                result.append(s.pop_item())

        return ("".join(result))
       
        
st = stack()
#st.InfixToPostfix("( A + B ) + 2 * A + ( X - Y )")
#st.InfixToPostfix("( A + B ) * ( C + D )")
#st.InfixToPostfix("( A + B ) * C")
#st.InfixToPostfix("( A + B ) * C - ( D - E ) * ( F + G )")
#st.InfixToPostfix(" 2 ^ ( A + B )")
#a + 2*X^(a+b)
#st.InfixToPostfix("A + 2 * X ^ ( A + B )")
#st.InfixToPostfix("( A + B ) * X")
st.InfixToPostfix("A+B*C/(E-F)")
```

```python
##Evaluate infix expression

class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
        
    def pop_item(self):
        return self.items.pop()
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)
    
    def PerformOperation(self,var1,var2,oper):
        
        if oper == "+":
            return var1 + var2
        if oper == "-":
            return var1 - var2
        if oper == "*":
            return var1 * var2
        if oper == "/":
            return var2 / var1
           
    def InfixEvaluation(self,exp):
        
        prec = {}
        
        prec["^"] = 4
        prec["*"] = 3
        prec["/"] = 3
        prec["+"] = 2
        prec["-"] = 2
        prec["("] = 1

        value = Stack()
        oper = Stack()
        l1 = exp.split(" ")
        
        for i in range(0,len(l1)):
                                   
            if l1[i] == " ":
                i += 1
                continue
            
            elif l1[i].isdigit():
                value.push(l1[i])
                
            elif l1[i] == "(":
                oper.push(l1[i])
            
            elif l1[i] == ")":
                
                while oper.peek() != "(":
                    var1 = float(value.pop_item())
                    var2 = float(value.pop_item())
                    temp = oper.pop_item()
                    newval = self.PerformOperation(var1,var2,temp)
                    value.push(newval)

                oper.pop_item()
                        
            elif l1[i] in prec:
                
                if oper.peek() == "(":
                    oper.push(l1[i])
                else:
                    
                    while not oper.is_empty() and prec[l1[i]] <= prec[oper.peek()]:
                        var1 = float(value.pop_item())
                        var2 = float(value.pop_item())
                        temp = oper.pop_item()
                        newval = self.PerformOperation(var1,var2,temp)
                        value.push(newval)
                        
                    oper.push(l1[i])          
                
        if not oper.is_empty():
            while not oper.is_empty():
                
                var1 = float(value.pop_item())
                var2 = float(value.pop_item())
                temp = oper.pop_item()
                newval = self.PerformOperation(var1,var2,temp)
                value.push(newval)
                   
        return round(newval,2)
                
                      
            
s = Stack()
#s.InfixEvaluation("10 * 2 * 6")
s.InfixEvaluation("100 * ( 2 + 12 / 4 ) / 14")

#s.InfixEvaluation("( 100 * 2 ) + ( 2 + 12 / 4 ) / 14")                
```

```python
### Maximum product of indexes of next greater on left and right
class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
        
    def pop_item(self):
        return self.items.pop()
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)
    
    def Product_of_Next_Greater_Element_On_Left_And_Right(self,arr):
        
        s = Stack()
        LeftIndex = Stack()
        RightIndex = Stack()
        s.push(arr[0])
        LeftIndex.push(0)
        index = 0
        while index <= len(arr):
            for i in range(index+1,len(arr)):
                if s.is_empty()==False:
                    if arr[i] > s.peek():
                        RightIndex.push(i)
                        s.pop_item()
       
            if s.is_empty()==False:
                RightIndex.push(0)
                s.pop_item()
            if index == len(arr)-1:
                break
            else:
                index += 1
                s.push(arr[index])
                
        #print("RightIndex:",RightIndex.items)
    
        index = 1
        s.push(arr[1])
        while  index <= len(arr):
            
            for i in reversed(range(index)):
                if s.is_empty() == False:
                                        
                    if arr[i] > s.peek():
                        
                        LeftIndex.push(i)
                        s.pop_item()
                        
            if s.is_empty()==False:
                LeftIndex.push(0)
                s.pop_item()
            if index == len(arr)-1:
                break
            else:
                index += 1
                s.push(arr[index])

        #print("LeftIndex:",LeftIndex.items)
        
        if LeftIndex.stack_size() == RightIndex.stack_size():
            
            output = [0] * LeftIndex.stack_size()
            ind = 0 
            while not LeftIndex.is_empty() and not RightIndex.is_empty():
                
                prod = LeftIndex.pop_item() * RightIndex.pop_item()
                output[ind] = prod
                ind += 1
            #print("output:",output)
        return max(output)
                   
s = Stack()
s.Product_of_Next_Greater_Element_On_Left_And_Right([1,1,1,0,1,1])
```

```python
## The Stock Span Problem
### The stock span problem is a financial problem where we have a series of n daily price quotes for a stock and
### we need to calculate span of stock’s price for all n days.

class Stack():
    
    def __init__(self):
        self.items = []
          
    def push(self,data):
        self.items.append(data)
        
    def pop_item(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return -1 
    
    def is_empty(self):
        return self.items == []
    
    def peek(self):
        
        if not self.is_empty():
            return self.items[-1]
        
    def stack_size(self):
        return len(self.items)
    
    def cal_stock_Price_span(self,price):
        
        n = len(price)
        span = [0] * n
        span[0] = 0
        s = Stack()
        s.push(0)
        print("stack:",s.items)
        for i in range(1,n):
            
            while not s.is_empty() and price[s.peek()] <= price[i]:
                s.pop_item()
                
            if s.is_empty():
                span[i] = i + 1
            else:
                span[i] = i - s.peek()

            s.push(i)

        return span
    
s = Stack()
s.cal_stock_Price_span([10,80,90,101,82])
            
```
