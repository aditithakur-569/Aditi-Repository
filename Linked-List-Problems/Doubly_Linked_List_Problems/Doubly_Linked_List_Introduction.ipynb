```python
## Doubly Linked list Implementation
## A node can be added in four ways
## 1) At the front of the DLL
## 2) After a given node.
## 3) At the end of the DLL
## 4) Before a given node.
## Start Date- 11-23-2020
## End Data - 11-23-2020
class node:
    
    def __init__(self, data,next = None, prev = None):
        
        self.data = data
        self.next = next
        self.prev = prev
        
class doubly_linked_list():
    
    def __init__(self,head = None):
        
        self.head = head
        self.size = 0
        
    
    def add_node_at_the_beginning(self,data):
        
        curr_node = self.head
        newnode = node(data)
        
        if not self.head:
            newnode.prev = self.head
            newnode.next = None
            self.head = newnode
        else:
            curr_node.prev = newnode
            newnode.next = curr_node
            self.head = newnode
            
    def add_node_at_the_end(self,data):
        
        curr = self.head
        newnode = node(data)
        if not curr:
            
            newnode.prev = self.head
            newnode.next = None
            self.head = newnode
        else:
            while curr.next != None:
                curr = curr.next
            curr.next = newnode
            newnode.prev = curr
            newnode.next = None
            
    def insert_node_after_given_node(self,gindex,data):
        
        self.cindex = 0
        curr = self.head
        newnode = node(data)  
        while self.cindex != gindex:
            curr = curr.next
            self.cindex += 1
            
        temp = curr.next
        curr.next = newnode
        newnode.prev = curr
        if curr.next != None:
            newnode.next = temp
        else:
            newnode.next = None
    
        
            
    def print_data_forward(self):
        
        curr = self.head
        while curr:
            print(curr.data, end = "-->")
            curr = curr.next
        print("None")
        
    def print_data_backward(self):
        
        curr = self.head
        while curr.next != None:
            curr = curr.next
            
        while curr != None:
            print(curr.data,end = "-->")
            curr = curr.prev
        print("None")
            
        
            
            

        

dl = doubly_linked_list()
dl.add_node_at_the_beginning(4)
dl.add_node_at_the_beginning(12)
dl.add_node_at_the_beginning(67)
dl.add_node_at_the_beginning(88)


#dl.print_data_backward()

dl.add_node_at_the_end(32)
dl.add_node_at_the_end(100)
#dl.print_data_forward()
#dl.insert_node_after_given_node(2,-1)
#dl.print_data_forward()
dl.print_data_forward()
```
