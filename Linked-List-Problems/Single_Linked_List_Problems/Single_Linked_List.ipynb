```python
#LinkedList
#Single LinkedList
## This program adds Node at the end of the list (add_node_at_the_end).
## This program adds Node at the beginning of the list (add_node_to_beginning).
## This program adds Node in between the nodes of the list (add_node_in_between_nodes) and also adds in the beginning
## or the end of the linked list.
## Calculates the size of the list (linked_list_size).
## Prints the list.
## Deletes the node by value and node can be either one of head, tail or body nodes.delete_node(self, val)
## Deletes the tail node.

class Node:
    
    def __init__(self,data):
        self.data = data
        self.nextnode = None       
    
class linkedlist:
    
    def __init__(self,head = None):
        self.head = head
        self.size = 0
    
    def linked_list_size(self):
        self.size = 0
        curr_node = self.head
        if self.head:
            while curr_node.nextnode != None:
                curr_node = curr_node.nextnode
                self.size += 1
                
            self.size += 1
        return self.size
    
    def add_node_to_beginning(self,data):
        newnode = Node(data)
        newnode.nextnode = self.head
        self.head = newnode
        self.size += 1
        return "Add Node Successful!"
        
    def add_node_at_the_end(self,data):
        
        new_node = Node(data)
        status = False
        if self.head:
            curr_node = self.head
            while curr_node.nextnode != None:
                curr_node = curr_node.nextnode
            curr_node.nextnode = new_node
            status = True
            
        else:
            ### To add the first node, if the list is empty
            new_node.nextnode = self.head
            self.head = new_node
            self.size += 1
            status = True
        if status == True: return "Add Node Successfull!"
        else: return "Add Node Unsuccessfull!"
    
    def add_node_in_between_nodes(self,data,index):
        
        llistlen = self.linked_list_size()
        ## Index provided by user to add the node
        self.index = index
        curr_node = self.head
        self.curr_pos = 0
        new_node = Node(data)
        self.status = False
        if llistlen == 0:
            print("List is empty. Operation can't be carried..")
            status = False
        else:
            if self.index > llistlen:
                print("Invalid Index!!!Node can't be added!!")
                status = False
            
            else:
                if self.index == 0:
                    new_node.nextnode = self.head
                    self.head = new_node
                    status = True
                else:
                    while self.curr_pos != self.index:
                        pre_node = curr_node
                        curr_node = curr_node.nextnode
                        self.curr_pos +=1
        
                    pre_node.nextnode = new_node
                    new_node.nextnode = curr_node
                    self.size += 1
                    status = True
           
        if status == True: return "Add Node Successfull!"
        else: return "Add Node Unsuccessfull!"
        
    def delete_end_node(self):
        
        curr_node =self.head 
        status = False
        if curr_node:
            while curr_node.nextnode != None:
                
                prev_node = curr_node
                curr_node = curr_node.nextnode 
            prev_node.nextnode = None
            status = True

        else:
            print("The list is empty !")
            status = False
        
        if status == True: return "Delete Node Successfull!"
        else: return "Delete Node Unsuccessfull!"
        
    def delete_node(self, val):
        self.val = val
        curr_node = self.head
        self.status = False
        if curr_node:
            if curr_node.data == self.val:
                curr_node = curr_node.nextnode
                self.head = curr_node
                self.status = True
            else:
                while curr_node.nextnode != None:
                    if curr_node.data != self.val:
                        prev_node = curr_node
                        curr_node = curr_node.nextnode
                        self.status = False
                    else:
                        curr_node = curr_node.nextnode 
                        prev_node.nextnode = curr_node
                        self.status = True
                else:
                    if ((curr_node.data == self.val) and (curr_node.nextnode == None)):
                        prev_node.nextnode = None
                        self.status = True
                    
                if self.status == True:print("Deletion Successful!!")
                else:print("Deletion Unsuccessful!! Data couldn't be found!")
        
        else:
            print("Empty List!!")
        
    
        

    
    def print_mylinkedlist(self):
        
        #mylist = []
        curr_node = self.head       
        while curr_node != None:            
            #mylist.append(curr_node.data)
            print(curr_node.data,end="-->")
            curr_node = curr_node.nextnode
        print(None)
        
           

l = linkedlist()
print(l.add_node_at_the_end('Gucci'))
print(l.add_node_at_the_end('Prada'))
print(l.add_node_at_the_end('Chanel'))
print(l.add_node_at_the_end('Urban Decay'))
print(l.add_node_at_the_end('Guess'))
print(l.add_node_at_the_end('Michael Kors'))
print(l.add_node_to_beginning('kids'))
print(l.add_node_to_beginning(5))
print(l.add_node_to_beginning(10034))
print(l.add_node_to_beginning('summer'))
print(l.add_node_to_beginning('SQL'))
print(l.add_node_in_between_nodes('Tommy Hilfigure',2))
print(l.add_node_in_between_nodes('POLO RALPH LAUREN',6))
print(l.add_node_in_between_nodes('Alfani',15))
print(l.add_node_to_beginning('Python'))
print(l.add_node_at_the_end('Reyansh'))
print(l.add_node_in_between_nodes('Alfani',15))
print(l.add_node_in_between_nodes('Armani',13))
print(l.add_node_to_beginning('MAC'))
print(l.add_node_to_beginning('Lu'))
print(l.add_node_at_the_end('Gucci'))
#print(l.add_node_at_the_end('Prada'))
#print(l.add_node_at_the_end('Chanel'))
#print(l.add_node_at_the_end('Urban Decay'))
#print(l.add_node_at_the_end('Guess'))
#print(l.add_node_at_the_end('Michael Kors'))
#print(l.add_node_at_the_end('T'))
#print(l.add_node_at_the_end('Louis Vuitton'))
#print(l.add_node_to_beginning('LEVIS'))
#print(l.add_node_at_the_end('Prada'))
#print(l.add_node_at_the_end('INC'))
#print(l.add_node_at_the_end('Jimmy Choo'))
#print(l.add_node_at_the_end('Gucci'))
#print(l.add_node_at_the_end('Nike'))
print(l.add_node_in_between_nodes('Chanel',2))
print(l.add_node_in_between_nodes('Armani',2))
#print(l.add_node_in_between_nodes('Alfani',0))
l.print_mylinkedlist()
```
