```python
###Give an algorithm for finding the sum of all elements in a binary tree.
## Recursive Approach - (Start date: 26th Jan,2021 -- End date: 27th Jan,2021)
## Iterative Approach - (Start date: 27th Jan,2021 -- End date: 27th Jan,2021)
##Sum of all the parent nodes having child node x(recur)(Start date: 27th Jan,2021 -- End date: 27th Jan,2021)
#Sum of all the parent nodes having child node x(iter)(Start date: 27th Jan,2021 -- End date: 27th Jan,2021)
import queue
class Node():
    
    def __init__(self,data):
        self.data = data
        self.LeftChild = None
        self.RightChild = None
        
class Binary_Tree():
    
    def __init__(self,root):
        self.root = Node(root)
        
    def insert_node_in_Level(self,root,data):
        
        q = queue.Queue()
        q.put(root)
        
        newnode = Node(data)
        while not q.empty():
           
            temp = q.get()
            if temp.LeftChild == None:
                temp.LeftChild = newnode
                return
            else:
                q.put(temp)
                
            if temp.RightChild == None:
                temp.RightChild = newnode
                return
            else:
                q.put(temp)
                
    def level_order_traversal(self,root):
        
        q = queue.Queue()
        q.put(root)
        q.put(None)
        
        while q.qsize() > 1:
            
            temp = q.get()
            if temp != None:
                print(temp.data,end= " ")
                if temp.LeftChild:
                    q.put(temp.LeftChild)

                if temp.RightChild:
                    q.put(temp.RightChild)
            else:
                q.put(None)
                print(end='\n')
                
    def cal_sum_of_all_nodes_recur(self,root,total = 0):
        
        if root == None:
            return 
 
        left_sum = self.cal_sum_of_all_nodes_recur(root.LeftChild)
        right_sum = self.cal_sum_of_all_nodes_recur(root.RightChild)
        
        if left_sum != None:
            total = total + left_sum
        if right_sum != None:
            total = total + right_sum
        return (total + root.data)
        
    def cal_sum_of_all_nodes_iter(self,root):
        
        q= queue.Queue()
        q.put(root)
        total = 0
        while q.qsize() >0:
            
            temp = q.get()
            total = total + temp.data
            if temp.LeftChild:
                q.put(temp.LeftChild)
            if temp.RightChild:
                q.put(temp.RightChild)
        return total
            
 

    def sum_of_parent_nodes_iter(self,root,x):
        
        q = queue.Queue()
        q.put(root)
        total = 0
        
        while q.qsize() > 0:
            
            temp = q.get()
            
            if temp.LeftChild:
                q.put(temp.LeftChild)
            if temp.RightChild:
                q.put(temp.RightChild)
            
           # if temp.LeftChild != None and temp.RightChild != None:
                
            #    if temp.LeftChild.data == x or temp.RightChild.data == x:
            #        total = total + temp.data 
            if ((temp.LeftChild != None and temp.LeftChild.data == x) or 
             (temp.RightChild != None and temp.RightChild.data == x)):
                total = total + temp.data
            
            
        return total
    
    def sum_of_parent_nodes_recur(self,root,x):
        
        total = [0]
        print(type(total))
        self.sum_of_parent_nodes_util(root,total,x)
        return total
        
        
    def sum_of_parent_nodes_util(self,root,total,x):
        
       
        if root == None:
            return
        
        
        if ((root.LeftChild != None and root.LeftChild.data == x) or 
             (root.RightChild != None and root.RightChild.data == x)):
                total[0] = total[0] + root.data
               
        self.sum_of_parent_nodes_util(root.LeftChild,total,x)
        self.sum_of_parent_nodes_util(root.RightChild,total,x)
    
        
        
        
                
            
    
           
            
            
   






bt = Binary_Tree(6)
bt.insert_node_in_Level(bt.root,4)
bt.insert_node_in_Level(bt.root,9)
bt.insert_node_in_Level(bt.root.LeftChild,4)
bt.insert_node_in_Level(bt.root.LeftChild,102)
bt.insert_node_in_Level(bt.root.RightChild,-1)
bt.insert_node_in_Level(bt.root.RightChild,100)
bt.insert_node_in_Level(bt.root.LeftChild.LeftChild,900)
bt.insert_node_in_Level(bt.root.RightChild.RightChild,4)
bt.insert_node_in_Level(bt.root.RightChild.RightChild,88)
bt.level_order_traversal(bt.root)
print("\n")
#bt.cal_sum_of_all_nodes_recur(bt.root)
#bt.cal_sum_of_all_nodes_iter(bt.root)
bt.sum_of_parent_nodes_iter(bt.root,4)
#bt.sum_of_parent_nodes_recur(bt.root,4)
```
