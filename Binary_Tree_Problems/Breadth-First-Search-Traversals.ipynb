```python
## Boundary Traversal of Binary Tree
## Start Date - End Date: 21st Jan - 21st Jan
class Node():
    
    def __init__(self,data):
        self.data = data
        self.LeftChild = None
        self.RightChild = None
        
class Binary_Tree():
    
    def __init__(self,root):
        self.root = Node(root)
        
    def insert_node_in_Level(self,root,data):
        
        q = queue.Queue()
        q.put(root)
        
        newnode = Node(data)
        while not q.empty():
           
            temp = q.get()
            if temp.LeftChild == None:
                temp.LeftChild = newnode
                return
            else:
                q.put(temp)
                
            if temp.RightChild == None:
                temp.RightChild = newnode
                return
            else:
                q.put(temp)
                
    def print_boundary(self,root):
        
        if root == None:
            return
        print(root.data,end = " ")
        
        self.print_boundary_left(root.LeftChild)
        self.print_leaf_nodes(root.LeftChild)
        self.print_leaf_nodes(root.RightChild)
        self.print_boundary_right(root.RightChild)
        
        
    def print_boundary_left(self,root):
        
       # if root:
        #    print("left-root:",root.data)
        #else:
         #   print("left-root:",None)
        
        if root:
            
            if root.LeftChild:
                print(root.data,end = " ")
                self.print_boundary_left(root.LeftChild)

            elif root.RightChild:
                print(root.data,end = " ")
                self.print_boundary_left(root.RightChild)

            
    def print_boundary_right(self,root):
        
        #if root:
        #    print("right-root:",root.data)
        #else:
        #    print("right-root:",None)
        
        if root:
           
            if root.RightChild:
                
                self.print_boundary_right(root.RightChild)
                print(root.data,end = " ")

            else:
                self.print_boundary_right(root.LeftChild)
                print(root.data,end = " ")
            
    def print_leaf_nodes(self,root):
        #if root:
         #   print("leaves:",root.data)
        #else:
         #   print("leaves:",None)
        
        if root:
            self.print_leaf_nodes(root.LeftChild)
        
            if root.LeftChild == None and root.RightChild == None:
                print(root.data,end= " ")
            
            self.print_leaf_nodes(root.RightChild)

bt = Binary_Tree(1)
bt.insert_node_in_Level(bt.root,2)
bt.insert_node_in_Level(bt.root,3)
bt.insert_node_in_Level(bt.root.LeftChild,4)
bt.insert_node_in_Level(bt.root.LeftChild,5)
bt.insert_node_in_Level(bt.root.LeftChild.RightChild,20)
bt.insert_node_in_Level(bt.root.LeftChild.RightChild,21)
bt.insert_node_in_Level(bt.root.RightChild,6)
bt.insert_node_in_Level(bt.root.RightChild,7)
bt.insert_node_in_Level(bt.root.RightChild.RightChild,90)
bt.insert_node_in_Level(bt.root.RightChild.RightChild,100)
bt.print_boundary(bt.root)
```

```python
##Spiral Order Traversal of Binary Tree
## Also called Zigzag Traversal
## Approach 1: Two stack Methods(start date:23rd Jan - -> End date: 23rd Jan)
## Approach 2: Deque(start date:23rd Jan - -> End date:23rd Jan)
## Reverse spiral order traversal(recursion)((start date:23rd Jan - -> End date:24thJan))
## Reverse spiral order traversal(iteration)((start date:24th Jan - -> End date:25thJan))

import queue
from collections import deque
class Node:
    
    def __init__(self,data):
        
        self.data = data
        self.LeftChild = None
        self.RightChild = None
        
class Binary_Tree:
    
    def __init__(self,root):
        
        self.root = Node(root)
          
    def insert_node(self,root,data):
        
        q = []
        q.append(root)
        
        while q:
            temp = q[0] 
            q.pop(0)
           
        newnode = Node(data)
        
        if temp.LeftChild == None:
            temp.LeftChild = newnode
            return
        else:
            q.append(temp)
            
        if temp.RightChild == None:
            temp.RightChild = newnode
            return
        else:
            q.append(temp)
            
    def insert_node_in_level(self,root,data):
        
        q = []
        q.append(root)
        
        newnode = Node(data)
        
        while q:
            temp = q.pop(0)
            if temp.LeftChild == None:
                temp.LeftChild = newnode
                return
            else:
                q.append(temp)
                
                
            if temp.RightChild == None:
                temp.RightChild = newnode
                return
            else:
                q.append(temp)
                
    def cal_height(self,root):
        
        if root == None:
            return 0
        
        l_height = self.cal_height(root.LeftChild)
        r_height = self.cal_height(root.RightChild)
        return max(l_height , r_height) + 1
                
        
            
    def level_order_traversal(self,root):
        
        q = []
        q.append(root)
        q.append(None)
        
        while len(q) > 1:
            
            temp = q.pop(0)
            
            if temp != None:
                print(temp.data,end= " ")
                
                if temp.LeftChild:
                    q.append(temp.LeftChild)
                if temp.RightChild:
                    q.append(temp.RightChild)
            else:
                q.append(None)
                print(end="\n")
                
    def spiral_order_traversal_stacks(self,root):
        
        ## Initializing stacks
        s1 = deque()
        s2 = deque()
        s1.append(root)
        
        while s1 or s2:
        
            while len(s1) != 0:

                curr = s1[0]
                if curr.LeftChild:
                    s2.append(curr.LeftChild)
                if curr.RightChild:
                    s2.append(curr.RightChild)

                print(curr.data,end=" ")
                s1.popleft()


            while len(s2) != 0:

                curr = s2[-1]                
                if curr.RightChild:
                    s1.appendleft(curr.RightChild)
                    
                if curr.LeftChild:
                    s1.appendleft(curr.LeftChild)
                
                
                print(curr.data,end=" ")
                s2.pop()
                
    def spiral_order_traversal_deque(self,root):
        
        d1 = deque()
        
        d1.append(root)
        d1.append(None)
        
        while len(d1) > 1:
            
            curr = d1[0]
            if curr != None:
                if curr.LeftChild:
                    d1.append(curr.LeftChild)
                if curr.RightChild:
                    d1.append(curr.RightChild)
                print(curr.data,end = " ")
                d1.popleft()
                    
            elif curr == None:
                print(end="\n")
                
                while d1[-1] != None:
                    curr = d1[-1]
                    print(curr.data,end=" ")
                    
                    if curr.RightChild:
                        d1.appendleft(curr.RightChild)
                    if curr.LeftChild:
                        d1.appendleft(curr.LeftChild)
                        
                    d1.pop()
                print(end="\n")
        #print()
        
    def reversed_spiral_order_traversal_recur(self,root):
        
        h = self.cal_height(root)

        for i in reversed(range(1,h+1)):
          
            if i % 2 == 0:
                flag = True
            else:
                flag = False
            self.print_reversed_spiral_order_traversal(root,i,flag)
            print(end = "\n")
            
            
    def print_reversed_spiral_order_traversal(self,root,level,flag):
        
        if root == None:
            return 

        if level == 1 :
            print(root.data,end=" ")
            return 

        if level > 1 and flag == True:
            
            self.print_reversed_spiral_order_traversal(root.LeftChild,level-1,flag)
            self.print_reversed_spiral_order_traversal(root.RightChild,level-1,flag)
                
                
        elif level > 1 and flag == False:
            
            self.print_reversed_spiral_order_traversal(root.RightChild,level-1,flag)
            self.print_reversed_spiral_order_traversal(root.LeftChild,level-1,flag)
            
            
    def reversed_spiral_order_traversal_iter(self,root):
            
            #q = queue.Queue()
            q = deque()
            d = deque()
           
            q.append(root)
            q.append(None)
            
            flag = False
            
            while len(q)>1:
                
                temp = q[0]    
                if temp != None:
                    
                    d.append(temp)
                    if temp.RightChild:
                        q.append(temp.RightChild)
                    if temp.LeftChild:
                        q.append(temp.LeftChild)
                        
                    q.popleft()
                          
                elif temp == None and flag == False:
                   
                    d.append(None)
                    q.append(None)
                    flag = True
                    q.popleft()
                        
                elif temp == None and flag == True:
                    d.appendleft(None)
                    d.append(None)
                    temp = q[-1]
                    while temp != None:
                        
                        d.appendleft(temp)
                        if temp.RightChild:
                            q.appendleft(temp.LeftChild)
                        if temp.LeftChild:
                            q.appendleft(temp.RightChild)
                        q.pop()
                        temp = q[-1]
                    flag = False

                    
            while len(d) > 1:
                
                if d[-1] != None:
                    while d[-1] != None:

                        print(d.pop().data,end=" ")
                    print(end="\n")
                    d.pop()
                    
                    
                if d[0] != None and len(d) > 1:

                    while d[0] != None:

                        print(d.popleft().data,end=" ")
                    print(end="\n")
                    d.popleft()
                    
            if d:
                print(d[0].data)
            
                    
                
                
                
                
                
                
                
                
        
                    
                    
                
                

      
                            
                        
                        
                    

            
            

 

            
                
                        
                
               
                
                
            
                
               
                
                
            
            
            
            
bt = Binary_Tree(1)
bt.insert_node(bt.root,2)
bt.insert_node(bt.root,3)
bt.insert_node(bt.root.LeftChild,4)
bt.insert_node(bt.root.LeftChild,5)
bt.insert_node(bt.root.RightChild,6)
bt.insert_node(bt.root.RightChild,7)
bt.insert_node(bt.root.LeftChild.LeftChild,8)
bt.insert_node(bt.root.LeftChild.LeftChild,9)
bt.insert_node(bt.root.LeftChild.RightChild,10)
bt.insert_node(bt.root.LeftChild.RightChild,11)
#bt.level_order_traversal(bt.root)
#print("\n")
#bt.spiral_order_traversal_stacks(bt.root)

#bt.spiral_order_traversal_deque(bt.root)
#print("\n")
#bt.reversed_spiral_order_traversal_recur(bt.root)
bt.reversed_spiral_order_traversal_iter(bt.root)


```
