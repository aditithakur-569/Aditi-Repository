```python
## Maximum element between two nodes of BST.[start-end: April 8th]

import queue
class Node:
    
    def __init__(self,data):
        
        self.data = data
        self.LeftChild = None
        self.RightChild = None
        

def level_order_traversal(root):
        
        q = queue.Queue()
        q.put(root)
        q.put(None)
        
        while q.qsize() > 1:
            
            temp = q.get()
            if temp != None:
                print(temp.data,end= " ")
                if temp.LeftChild:
                    q.put(temp.LeftChild)

                if temp.RightChild:
                    q.put(temp.RightChild)
            else:
                q.put(None)
                print(end='\n')
                
                
def insert_node_iter(root,val):

    if root == None:
        root = Node(val) 

    if val > root.data:
        root.RightChild = insert_node_iter(root.RightChild,val)
    
    elif val < root.data:
        root.LeftChild = insert_node_iter(root.LeftChild,val)
    return root


def find_maximum_element_between_nodes(arr,a,b):
    
    root = Node(arr[0])
    n = len(arr)
    global Max1,Max2
    Max1,Max2 = -99999,-99999
    for i in range(1,n):
        insert_node_iter(root,arr[i])
    level_order_traversal(root)
    print("\n")
    LCA = find_LCA(root,a,b)
    x = find_maximum_element(LCA,a,b)
    print("The maximum element between two nodes is:",x)
    
def find_LCA(root,a,b):
    
    if root == None:
        return
    
    if a < root.data and b < root.data:
        return find_LCA(root.LeftChild,a,b)
    if a > root.data and b > root.data:
        return find_LCA(root.RightChild,a,b)
    
    return root

    
def find_maximum_element(root,a,b):
    
    global Max1,Max2
    if root == None:
        return
    
    curr1, curr2 = root,root
    
    while curr1.data != a:
        
        Max1 = max(Max1,curr1.data)
        if a < curr1.data:
            curr1 = curr1.LeftChild
        elif a > curr1.data:
            curr1 = curr1.RightChild

        
    while curr2.data != b:
        
        Max2 = max(Max2,curr2.data)
        if b < curr2.data:
            curr2 = curr2.LeftChild
        elif b > curr2.data:
            curr2 = curr2.RightChild

    return max(Max1,Max2)
        

    
find_maximum_element_between_nodes([18,36,9,6,12,10,1,8],1,36)
```
