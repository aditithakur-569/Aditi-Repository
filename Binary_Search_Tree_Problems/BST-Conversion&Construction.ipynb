```python
#Merge two BSTs with limited extra space.[start-end: April 15th]

import queue
class Node:
    def __init__(self,data):
        
        self.data = data
        self.LeftChild = None
        self.RightChild = None
        
def insert_node_iter(root,data):
    
    if root == None:
        root = Node(data)
        return root
        
    curr = root
    parent = None
    while curr != None:
        
        parent = curr
        if data < curr.data:
            curr = curr.LeftChild
            
        elif data > curr.data:
            curr = curr.RightChild
            
    if data < parent.data:
        parent.LeftChild = Node(data)
    elif data > parent.data:
        parent.RightChild = Node(data)
        
    return root

def level_order_traversal(root):
        
        q = queue.Queue()
        q.put(root)
        q.put(None)
        
        while q.qsize() > 1:
            
            temp = q.get()
            if temp != None:
                print(temp.data,end= " ")
                if temp.LeftChild:
                    q.put(temp.LeftChild)

                if temp.RightChild:
                    q.put(temp.RightChild)
            else:
                q.put(None)
                print(end='\n')
                
def inorder_traversal(root):
    
    if root == None:
        return 
    
    inorder_traversal(root.LeftChild)
    print(root.data,end=" ")
    inorder_traversal(root.RightChild)
                
def merge_two_bsts(root1,root2):
    
    ##initializing stacks
    s1,s2 = [],[]
    curr1, curr2 = root1, root2
    
    if curr1 == None:
        return inorder_traversal(root2)
    
    if curr2 == None:
        return inorder_traversal(root1)
    
    while s1 or s2 or curr1 or curr2:
        
        if curr1 or curr2:
        
            while curr1:
                s1.append(curr1)
                curr1 = curr1.LeftChild

            while curr2:
                s2.append(curr2)
                curr2 = curr2.LeftChild
                
        else:
            
            if len(s2) == 0:
                while s1:
                    curr1 = s1.pop(-1)
                    curr1.LeftChild = None
                    inorder_traversal(curr1)
                    return
                    
            if len(s1) == 0:
                
                while s2:
                    curr2 = s2.pop(-1)
                    curr2.LeftChild = None
                    inorder_traversal(curr2)
                    return
                    
        
        
        curr1 = s1.pop(-1)
        curr2 = s2.pop(-1)
        
        if curr1.data < curr2.data:
            print(curr1.data,end=" ")
            curr1 = curr1.RightChild
            s2.append(curr2)
            curr2 = None
            
        elif curr2.data < curr1.data:
            print(curr2.data,end=" ")
            curr2 = curr2.RightChild
            s1.append(curr1)
            curr1 = None
            
        elif curr1.data == curr2.data:
            
            print(curr1.data,end=" ")
            curr1 = curr1.RightChild
            curr2 = curr2.RightChild
                    
                    

                
root1 = Node(25)
insert_node_iter(root1,3)
insert_node_iter(root1,7)
insert_node_iter(root1,2)
insert_node_iter(root1,30)
level_order_traversal(root1)
print("\n")
root2 = Node(25)
insert_node_iter(root2,8)
insert_node_iter(root2,28)
insert_node_iter(root2,21)
insert_node_iter(root2,45)
level_order_traversal(root2)
print("\n")

merge_two_bsts(root1,root2)
```
