```python
from queue import Queue  
  
# Queue elements after sortedIndex are  
# already sorted. This function returns  
# index of minimum element from front to  
# sortedIndex  
def minIndex(q, sortedIndex): 
    min_index = -1
    min_val = 999999999999
    n = q.qsize()
   # print("minIndex...")
    print("sortedIndex:",sortedIndex)
    for i in range(n): 
        curr = q.queue[0]  
        #q.get()
        q.get()
        print("i:",i,"curr:",curr,"min-val:",min_val,"queue:",q.queue)
        if (curr <= min_val and i <= sortedIndex): 
            min_index = i  
            min_val = curr 
        q.put(curr) 
        print("q after put:",q.queue,"min-val:",min_val)       # C++ STL 
    return min_index 
  
# Moves given minimum element to  
# rear of queue  
def insertMinToRear(q, min_index):
    #print("\n")
    #print(("insertMinToRear",q.queue))
    #print("min-index:",min_index)
    min_val = None
    n = q.qsize() 
    for j in range(n): 
        curr = q.queue[0]  
        q.get() 
        #print("j:",j,"curr:",curr,"q:",q.queue)
        if (j != min_index):  
            q.put(curr)
            print("put-q:",q.queue)
        else: 
            min_val = curr 
    q.put(min_val)
    #print("q after inserting min to rear:",q.queue)
    #print("\n")

def sortQueue(q): 
    for i in range(1, q.qsize() + 1): 
        print("i:",i,"Main Queue:",q.queue)
        min_index = minIndex(q, q.qsize() - i)  
        insertMinToRear(q, min_index) 
  
# Driver code  
if __name__ == '__main__': 
    q = Queue() 
    #q.put(30)  
    #q.put(11)  
    #q.put(15)  
    #q.put(4)
    q.put(3) 
    q.put(21) 
    q.put(5) 
    q.put(7) 
    q.put(9) 
    print(q.queue)  
    # Sort queue  
    sortQueue(q)  
      
    # Prsorted queue  
    while (q.empty() == False): 
        print(q.queue[0], end = " ")  
        q.get() 
```
