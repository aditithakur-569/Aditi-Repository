```python
##Given an array and an integer k, find the maximum for each and every contiguous subarray of size k.
#Algorithm:

#1) At i= 0, store the first index, which is 0 in Deque.
#2) From i = 1,
#    a) If Deque length >0 and d[0] <= i-k, remove front of the Deque.[This step is easy to understand with the
## test case ([1,3,1,2,0,5],3) 
#    a) if arr[Deque[Last]] > arr[i], do nothing.
#    b) elif arr[Deque[Last]] < arr[i], remove last index from Deque.
    
#3) Add the index to the Deque.
#4) if i >= k-1, append the arr[Deque[front]] to the output.

from collections import deque

def MaxSlidingWindow(arr,k):
    
    d = deque()
    output = []
    d.append(0)
    if len(arr)==1:
        return arr
    elif k == 1:
        return arr

    else:
        
        for i in range(1,len(arr)):

            if len(d)> 0 and d[0] <= i-k:
                d.popleft()

            while len(d) >0 and arr[d[-1]] < arr[i]:
                d.pop()
        
            d.append(i)

            if i >= k-1:    
                output.append(arr[d[0]])
        return output
            
            
        
#MaxSlidingWindow([1, 2, 3, 1, 4, 5, 2, 3, 6],3)        
#MaxSlidingWindow([6,0,-1,22,4,37,3,7,99],3)
#MaxSlidingWindow([1,-1],1)
#MaxSlidingWindow([1,3,1,2,0,5],3)
#MaxSlidingWindow([-3,3,2,6,5,2,-4],4)
#MaxSlidingWindow([7,2,4],2)
MaxSlidingWindow([9,0,7,2,4,-1,19],2)         
```

```python
##Find the first circular tour that visits all petrol pumps..
## Naive Method....
## Start Date: 10-Dec-2020
## End Date: 10-Dec-2020
## Optimized Method (Solution accepted on leetcode

from collections import deque

def petrol_pump(gas,dist):
    
    q = deque()
    surplus = 0
    deficit = 0
    
    if len(gas) != len(dist):
        return -1
    else:
        for i in range(0,len(gas)):
            
            if surplus + gas[i] >= dist[i]:
                q.append(i)
                surplus = (surplus + gas[i]) - dist[i]
            else:
                deficit += surplus + gas[i] - dist[i]
                surplus = 0
                if len(q) > 0:
                    q.clear()
        if surplus + deficit >=0:
            return q.popleft()
        else:
            return -1
    
                
        
#petrol_pump([4,6,7,4],[6,5,3,5])
#petrol_pump([1,2,3,4,5],[3,4,5,1,2])
#petrol_pump([2,3,4],[3,4,3])
#petrol_pump([4,2,6,7,9],[5,6,7,3,2])
#petrol_pump([5,1,2,3,4],[4,4,1,5,1])
#petrol_pump([7,1,0,11,4],[5,9,1,2,5])
#petrol_pump([2],[2])
#petrol_pump([100,101,102,103,104,105,106,107,108,109],[121,22,34,5,6,7,9,600,324,699])           
    
petrol_pump([50,30,40,10,8,45,30,20,10,81],
           [45,20,50,12,30,55,45,25,15,1])   
    
```
