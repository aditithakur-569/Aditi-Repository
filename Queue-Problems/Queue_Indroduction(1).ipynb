```python
### Very Important Question
##How to efficiently implement k Queues in a single array?
## Create a data structure kQueues that represents k queues. Implementation of kQueues should use only one array,
## i.e.,k queues should use the same array for storing elements. Following functions must be supported by kQueues.

##enqueue(int x, int qn) –> adds x to queue number ‘qn’ where qn is from 0 to k-1
##dequeue(int qn) –> deletes an element from queue number ‘qn’ where qn is from 0 to k-1
##Start Date: 30-Nov-2020 
## End Date: 1-Dec-2020

class KQueues():
    
    def __init__(self,arrsize,N):
        ## arrsize: Size of Arry, N: Number of Queues to implement in the array
        self.arrsize = arrsize
        self.MainArr = [0] * arrsize
        self.N = N
        self.front = [-1] * N
        self.rear = [-1] * N
        self.Next = [0] * arrsize
        self.free = 0
        
        for i in range(0,arrsize):
            self.Next[i] = i + 1
        self.Next[i] = -1
        #print("Next:",self.Next)
    def is_Queue_Empty(self,Qn):
        if self.front[Qn] == -1:
            return True
        else:
            return False
        
    def is_array_full(self):
        return self.free == -1
        
        
    def enqueue(self,Qn,data):
        
        # Qn - -> Queue number
        if self.is_array_full():
            print("Space unavalaible...Queue Overflow..")
            return
        else:
            
            i = self.free
            self.free = self.Next[i]
            if self.front[Qn]== -1:
                self.front[Qn] = i

            else:
                self.Next[self.rear[Qn]] = i
        self.rear[Qn] = i
        self.Next[i] = -1
        self.MainArr[i] = data
        
        
    def dequeue(self,Qn):
        
        if self.is_Queue_Empty(Qn):
            print("Empty Queue...")
            return
        else:
            i = self.front[Qn]
            self.front[Qn] = self.Next[i]
            self.Next[i] = self.free
            self.free = i

 
        
q = KQueues(6,3)

q.enqueue(0,3)
q.enqueue(1,4)
q.enqueue(2,12)
q.enqueue(0,90)
q.dequeue(0)
q.enqueue(0,-89)
q.enqueue(1,77)
q.enqueue(0,66)
q.dequeue(0)
q.dequeue(0)
q.dequeue(0)
print(q.MainArr)          
            
            
        

```

```python
## Reversing a queue using recursion..
## Given a queue, write a recursive function to reverse it.
## Standard operations allowed :
## enqueue(x) : Add an item x to rear of queue.
## dequeue() : Remove an item from front of queue.
## empty() : Checks if a queue is empty or not.
## Start Date/End Date = 1-Dec-2020

class Queue:
    
    def __init__(self,N):
        
        self.N = N
        self.queue = [0] * N
        self.front = -1
        self.rear = -1
        self.free = 0
        
    def is_empty(self):
        return (self.front == -1)
        
        
    def enqueue(self,data):
        self.data = data
        
        if (self.rear +1) % self.N == self.front:
            print("Queue Overflow..")
            return 
        else:
            if self.front == -1:
                self.front = self.rear = 0
            else:
                self.rear = (self.rear + 1) % self.N
                
            self.queue[self.rear] = self.data
            
    def dequeue(self):
        
        if self.front == -1:
            print("Empty Queue")
            return
        elif self.front == self.rear:
            self.front = self.rear = -1
        else:
            
            self.front = (self.front + 1) % self.N
            
    def reverse_queue(self):
        
        if self.is_empty():
            return
        else:
            val = self.queue[self.front]
            self.dequeue()
            self.reverse_queue()
            self.enqueue(val)
            
            

q = Queue(5)
q.enqueue(12)
q.enqueue(-1)
q.enqueue(89)
q.enqueue(72)
q.enqueue(75)
#q.enqueue(100)
#q.enqueue(9)
#q.enqueue(43)
#q.enqueue(999)
#q.enqueue(67)
#q.enqueue(2)
print(q.queue)
q.dequeue()
q.dequeue()
#print(q.queue)
#q.dequeue()
#q.dequeue()
#q.dequeue()
#q.dequeue()
q.reverse_queue()
#q.dequeue()
q.enqueue(2)
q.enqueue(8)
print(q.queue)
q.reverse_queue()
print(q.queue)
        
        
```

```python
##Find the first circular tour that visits all petrol pumps..
## Naive Method....
## Start Date: 9-Dec-2020
## End Date: 10-Dec-2020
## Naive Method (29/31 test cases passed on leetcode.....)
## Optimized Method: Using Deque--See Deque

def petrol_pump(gas,dist):
    
    surplus = 0
    temp = []
    index = -1
    found = False
    if len(gas) != len(dist):
        return -1
    else:
        for i in range(0,len(gas)):
           
            if (surplus + gas[i]) >= dist[i]:
                if found == False:
                    index = i
                    found = True
                surplus = (surplus + gas[i]) - dist[i]
                
            else:
                if found == True:
                    while index!=i:
                        temp.append(index)
                        index +=1
                temp.append(i)
                found = False
                surplus = 0

        while temp:
            var = temp.pop(0)
            if (surplus + gas[var]) >= dist[var]:
                
                surplus = (surplus + gas[var]) - dist[var]
                found = True
            else:
                index = -1
        return index
    
#petrol_pump([4,6,7,4],[6,5,3,5])
#petrol_pump([1,2,3,4,5],[3,4,5,1,2])
#petrol_pump([2,3,4],[3,4,3])
#petrol_pump([4,2,6,7,9],[5,6,7,3,2])
#petrol_pump([5,1,2,3,4],[4,4,1,5,1])

#petrol_pump([7,1,0,11,4],[5,9,1,2,5])
#petrol_pump([2],[2])
```
